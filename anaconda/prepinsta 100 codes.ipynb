```python
#prime numbers in a given range
r=int(input("Enter upper limit: "))
for a in range(2,r+1):
    k=0
    for i in range(2,a//2+1):
        if(a%i==0):
            k=k+1
    if(k<=0):
        print(a)
```

```python
#Sum of digits of a number
n=int(input())
s=0
while(n!=0):
    r=n%10
    s+=r
    n=n//10
print(s)
```

```python
#Reverse of a number
n=int(input())
s=0
while(n!=0):
    r=n%10
    s=s*10+r
    n=n//10
print(s)
```

```python
#Palindrome
n=int(input())
temp=n
s=0
while(n!=0):
    r=n%10
    s=s*10+r
    n=n//10
if(temp==s):
    print("Palindrome")
else:
    print("Not Palindrome")
```

```python
#Armstrong
n=int(input())
temp=n
s=0
while(n!=0):
    r=n%10
    s=s+(r*r*r)
    n=n//10
if(temp==s):
    print("Armstrong")
else:
    print("Not Armstrong")
```

```python
#Strong Number
temp=n
s=0
while(n!=0):
    r=n%10
    f=1
    for i in range(1,r+1):
        f=f*i
    s=s+f
    n=n//10
if(temp==s):
    print("Strong")
else:
    print("Not")
```

```python
#Strong 2nd method 
def fact(n):
    f=1
    for i in range(1,n+1):
        f=f*(i)
    return f
def strong(n):
    s=0
    while(n!=0):
        r=n%10
        s=s+fact(r)
        n=n//10
    return s
n=int(input())
temp=n
b=strong(n)
if(b==n):
    print("Strong")
else:
    print("Not")
```
