```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("bikebuyer1.csv")
```

```python
df.isnull().sum()
```

```python
df["ID"].fillna(df["ID"].mean(), inplace=True)
marr_mode =df["Marital Status"].mode()
df["Marital Status"].fillna(marr_mode[0], inplace=True)
gender_mode = df["Gender"].mode()
df["Gender"].fillna(gender_mode[0], inplace=True)
df["Children"].fillna(df["Children"].mean(), inplace=True)
df["Commute Distance"].fillna(df["Commute Distance"].mean(), inplace=True)
```

```python
df.isnull().sum()
```

```python
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
df["Bike Buyer"]= le.fit_transform(df["Bike Buyer"])
```

```python
df.head()
```

```python
x = df.iloc[:,:12].values
y = df.iloc[:,12].values
x
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
```

```python
ct = ColumnTransformer([('oh', OneHotEncoder(), [1,2,5,6,7,10])], remainder = 'passthrough')
x_e = ct.fit_transform(x)
```

```python
x_e
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train, y_test= train_test_split(x_e,y,test_size=0.3, random_state=0)
```

```python
x_train.shape,x_test.shape,y_train.shape, y_test.shape
```

```python
from sklearn.linear_model import LogisticRegression
log = LogisticRegression()
```

```python
log.fit(x_train, y_train)
```

```python
log_pred = log.predict(x_test)
```

```python
log_pred
```

```python
y_test
```

```python
from sklearn.metrics import accuracy_score
```

```python
accur = accuracy_score(y_test, log_pred)
```
