```python
n=int(input())
arr=[]
for i in range(n):
    arr.append(str(input()))
for i in arr:
    num=arr.count(i)
    if(num%2==0):
        print(i,)
        break;
    else:
        print("-1")
```

```python
n= int(input())
l = []
for i in range(n):
    e = int(input())
    l.append(e)
l1 = set(l)
for i in l1:
    m = l.count(i)
    if(m%2==0):
        print(i)
        break
    else:
        print(-1)
```
