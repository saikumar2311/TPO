```python
#amazon
#Return the index of the item which is distinct 
#you have to return index of the first distinct
s="sasameer"
s1=list(s)
d={}
for i in s1:
    if(i in d):
        d[i]+=1
    else:
        d[i]=1
l=list(d.keys())
y=0
flag=0
for i in l:
    if(d.get(i)==1):
        flag=1
        y=i
        break
    else:
        flag=0
if(flag==1):
    print(s1.index(y)+1)
else:
    print("-1")
```
